```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Crear un grafo no dirigido
G = nx.Graph()

# Agregar nodos uno a uno
G.add_node("A")

# Agregar nodos en una lista
G.add_nodes_from(["B", "C", "D"])

# Agregar aristas uno a uno
G.add_edge("A", "B")
G.add_edge("A", "C")

# Agregar aristas en una lista
G.add_edges_from([("B", "C"), ("C", "D")])

nx.draw(G, with_labels=True, node_color="lightblue", font_weight="bold")
```

```python
# Crear un DataFrame de ejemplo
data = {
    "Fuente": ["A", "A", "B", "C", "D", "E"],
    "Destino": ["B", "C", "D", "D", "E", "A"],
    "Peso": [1, 2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Crear un grafo desde el DataFrame
G = nx.from_pandas_edgelist(df, source="Fuente", target="Destino", edge_attr="Peso")

# Dibujar el grafo
pos = nx.spring_layout(G)  # Posiciones de los nodos
nx.draw(G, pos, with_labels=True, node_color="lightblue", font_weight="bold")
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, "Peso"))
plt.show()
```
